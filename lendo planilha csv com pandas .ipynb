```python
import pandas as pd 



#às vezes precisaremos mudar o encoding. Possiveis valores para testar:
#encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

vendas_df= pd.read_csv('Contoso - Vendas - 2017.csv', sep=';')
cadastro_produtos_df= pd.read_csv('Contoso - Cadastro Produtos.csv', sep=';')
#promocoes_df= pd.read_csv('Contoso - Promocoes.csv', sep=';')
clientes_df= pd.read_csv('Contoso - Clientes.csv', sep=';')
lojas_df= pd.read_csv('Contoso - Lojas.csv', sep=';')

#usamos o display para exibir varios data frames

display(vendas_df)
display(cadastro_produtos_df)
#display(promocoes_df)
display(clientes_df)
display(lojas_df)
```

-

```python
#removendo colunas com .drop(['',''],) *lembrando que por padrao ele exclui linhas, para excluir colunas colocar
# axis =1

clientes_df = clientes_df.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9','Unnamed: 10'], axis=1)
display(clientes_df)
```

```python
#atraves de uma lista [['...', '...']] colocando o nome das colunas desejadas eu crio um novo dataframes 

clientes_df = clientes_df[['ID Cliente', 'E-mail']]
cadastro_produtos_df = cadastro_produtos_df[['ID Produto', 'Nome do Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]
display(cadastro_produtos_df)
```

```python
# juntando todas as colunas desejadas para obter apenas 1 data frames 

vendas_df = vendas_df.merge(cadastro_produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')

display(vendas_df)
```

```python
#renomeando coluna em forma de dicionario  

vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail do Cliente'})
display(vendas_df)
```

- O 1º passo de toda Análise de Dados é você entender o que existe na sua base de dados

Usaremos o .info() para isso

```python
### Qual cliente que comprou mais vezes?

# Usaremos o método .value_counts() para contar quantas vezes cada valor do Dataframe aparece
# Usaremos o método .plot() para exibir um gráfico, e um meio de uma tupla (método=( x,y)) escolhemos o tamanho do grafico 

maiores_compras = vendas_df['E-mail do Cliente'].value_counts()
display(maiores_compras)
maiores_compras[:5].plot(figsize=(15,5), yticks=range(68,80,1)) # yticks=range(68,100,1) determina a sequência
```

```python
vendas_df.info()
lojas_df.info()
promocoes_df.info()
clientes_df.info()
cadastro_produtos_df.info()
```

```python
### Qual a Loja que mais vendeu?

# Usaremos o .groupby para agrupar o nosso dataframe, de acordo 
# com o que queremos .sum (somando as quantidades de vendas, por exemplo) 

#.sort_values('Quantidade Vendida', ascending=False) para ordenar do maior para o menor, para escolhar o modelo do grafica kind=' modelo '

loja_com_maiores_vendas = vendas_df.groupby('Nome da Loja').sum() #a coluna que agrupa vem como indice no display, por isso na proxima visualizaçõa não precisa ter passada
loja_com_maiores_vendas = loja_com_maiores_vendas.sort_values('Quantidade Vendida', ascending=False)
loja_com_maiores_vendas = loja_com_maiores_vendas[['Quantidade Vendida']]
loja_com_maiores_vendas[:5].plot(figsize=(15,5), kind='bar') 
display(loja_com_maiores_vendas)
```

```python
# pegando o maior valor e melhor loja de outra forma min/max

maior_valor = loja_com_maiores_vendas['Quantidade Vendida'].max()
melhor_loja = loja_com_maiores_vendas['Quantidade Vendida'].idxmax()
print(maior_valor, melhor_loja)

#menor valor
loja_com_maiores_vendas[-1:]
```

```python
# Primeiro, vamos aplicar uma função normalmente. Qual o % das vendas que foi devolvido?

#Para isso vamos somar as quantidades nas colunas correspondentes. Lembrando, o % vai ser: Total Devolvido / Total Vendido.

quant_vend = vendas_df[ 'Quantidade Vendida'].sum()
quant_devol = vendas_df['Quantidade Devolvida'].sum()

print('{:.2%}'.format(quant_devol / quant_vend))
```
