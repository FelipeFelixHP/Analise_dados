### Importando módulos e arquivos 


Temos os dados de 2019 de uma empresa de prestação de serviços. 

- CadastroFuncionarios
- CadastroClientes
- BaseServiçosPrestados

```python
import pandas as pd 

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',' )
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')


#retirando colunas, por que achei a informações desnecessaria para a análise 

funcionarios_df = funcionarios_df.drop(['Estado Civil'], axis=1)



display(funcionarios_df)
display(clientes_df) #informações unica / tabela caracteristicas 
display(servicos_df) #informações repetidas / tabela fato 
```

### Valor Total da Folha Salarial ->

```python
FolhaSalarial = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']

display(FolhaSalarial)


funcionarios_df['Folha Salarial total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print('Total da Folha Salarial é de R${:,}'.format(sum(funcionarios_df['Folha Salarial total'])))
display(funcionarios_df)
```

### Qual foi o faturamento da empresa?

```python
#poderia ser também 

# faturamento_total = servicos_df[].merge(clientes_df[], on='ID Cliente') 

#o parametro on='' é a coluna que liga as duas tabelas 

#Só que desse jeito iria todas as colunas na geração do novo data frame

faturamento_total = servicos_df[['ID Cliente','Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente','Valor Contrato Mensal' ]]) 
faturamento_total['Valor total do Contrato'] = faturamento_total['Tempo Total de Contrato (Meses)'] * faturamento_total['Valor Contrato Mensal']
#sempre vai juntar a tabela de caracteristicas com a fabela fato, você pega a informações da tabela caracteristicas e joga na tabela fato

print    ('O faturamento da empreda foi de R${:,}'.format(sum(faturamento_total['Tempo Total de Contrato (Meses)'] * faturamento_total['Valor Contrato Mensal'])))

display(faturamento_total)
```

### Qual o % de funcionários que já fechou algum contrato?

```python
funcionarios_fecharam_contrato = len(servicos_df['ID Funcionário'].unique())
funcionarios_total = len(funcionarios_df['ID Funcionário'])
print('Porcentagem dos Funcionários que fecharam  contrato {:.2%}'.format(funcionarios_fecharam_contrato / funcionarios_total))
display(funcionarios_total)
display(funcionarios_fecharam_contrato)
```

###  Calcule o total de contratos que cada área da empresa já fechou

```python
areas_da_empresa = funcionarios_df['Area'].unique()
display(areas_da_empresa)

quantidade_de_contratos_por_areas = servicos_df[['ID Funcionário', 'Tempo Total de Contrato (Meses)']].merge(funcionarios_df[['ID Funcionário', 'Area']])
display(quantidade_de_contratos_por_areas)

total_por_areas = quantidade_de_contratos_por_areas['Area'].value_counts()
display(total_por_areas)

total_por_areas.plot(kind='bar')
```

### Calcule o total de funcionários por área

```python
total_funcionario = funcionarios_df['Area'].value_counts()
display(total_funcionario)

total_funcionario.plot(kind='pie')
```

 ### Qual o ticket médio mensal (faturamento médio mensal) dos contratos?

```python
ticket_médio = clientes_df['Valor Contrato Mensal'].mean()
print(ticket_médio)
```
